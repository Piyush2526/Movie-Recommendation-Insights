```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("movie_dataset.csv")

# Set style
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="genre", hue="genre", palette="viridis", legend=False)
plt.title("Number of Movies by Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()

```

```python
yearly_rating = df.groupby("year")["rating"].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=yearly_rating, x="year", y="rating")
plt.title("Average Movie Rating Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="rating", y="revenue_million", hue="genre")
plt.title("Rating vs Revenue")
plt.xlabel("Rating")
plt.ylabel("Revenue (Million USD)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="genre", y="runtime_min", hue="genre", palette="Set2", dodge=False)
plt.title("Runtime Distribution by Genre")
plt.xlabel("Genre")
plt.ylabel("Runtime (Minutes)")
plt.legend([],[], frameon=False)  # hide duplicate legend
plt.show()
```

```python
plt.figure(figsize=(10,6))
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["rating"], bins=20, kde=True, color="purple")
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```
